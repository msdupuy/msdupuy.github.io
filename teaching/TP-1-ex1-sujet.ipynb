# Schemes and initial data

```python
def u01(x):
    return np.sin(2*np.pi*x)

def u02(x):
    y = np.mod(x,1)
    if y < 0.25:
        return 0.0
    elif y < 0.75: 
        return 1.0
    else:
        return 0.0

#exact solution
def u_sol(x,t,a,u0):
    return u0(x-a*t)
```

```python
x = np.linspace(0,1,100)
y01 = [u01(z) for z in x]
y02 = [u02(z) for z in x]
fig, ax = plt.subplots()
ax.plot(x,y01,label="u_01")
ax.plot(x,y02,label="u_02")
plt.legend()
plt.show()
```
